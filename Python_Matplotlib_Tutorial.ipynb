```python
x = ['Mon','Tues','Wed','Thur','Fri']
y = [10,15,7,20,12]

plt.plot(x,y)

plt.title('Crockery Sales this week')
plt.xlabel('Day of the week')
plt.ylabel('Sales Per Day')

plt.show()
```

```python
#plt.plot(x,y,color='color name', linestyle='line_style',linewidth=value, marker='marker symbol', label= 'label name')
months = [1,2,3,4,5]
sales = [3000, 4000, 5000, 6000, 10000]

plt.plot(months, sales, color = 'red', linestyle = '--', linewidth= 3, marker = 'o', label = '2025 Sales Data of Crockery')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.title('Monthly Sales Data Report')
plt.legend(loc='upper left', fontsize=12)
plt.xticks([1,2,3,4,5], ['M1','M2','M3','M4','M5'])
plt.grid(color='gray', linestyle=':',linewidth=1)
plt.show()
```

# Bar Chart


```python
#plt.bar(x, height, color='color name', width = value, label='label name')
```

```python
product = ['A', 'B', 'C', 'D']
sales = [100,1500, 300, 2000]

plt.bar(product, sales, color='pink', label='Sales 2025')
plt.title('Product sales of 2025')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.grid(color = 'lightgray')
plt.legend(loc='upper left', fontsize = 12)
plt.show()
```

# Pie Chart

```python
#plt.pie(values, labels=label_list, color=color_list, autopct='%1.1f%%')

country = ['India', 'America', 'Bangladesh', 'South Africa']
Sales = [20000, 30000, 40000,50000]
plt.pie(Sales, labels=country, autopct='%1.1f%%', colors=['red', 'lightblue', 'lightgreen', 'coral'])
plt.title('Sales As Per Country')
plt.show()
```

# Histogram
used for continuous data in which one or more data gets repeated

```python
#plt.hist(data, bins = number_of_bins, color='colorname', edgecolor='black')

scores = [45,67,89,56,78,88,92,60,74,81,59,66,75,82,70,73,60,68,77]

plt.hist(scores, bins=5, color='cyan', edgecolor='black')
plt.xlabel('Score Range')
plt.ylabel('Number of students')
plt.title('Score Range Of Students')
plt.show()
```

# Scatter Plots 


```python
#plt.scatter(x,y, color='color_name', marker='marker_style', label='label name')
```

```python
hours_studied = [1,2,3,4,5,6,7,8]
exam_scores = [50,55,60,70,75,80,85,95]

plt.scatter(hours_studied, exam_scores, color='green', marker='o', label='Student Data')
plt.title('Relationship between studytime and exam scores')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```

```python
plt.scatter([1,2,3], [50,55,60], color = 'blue', label='Class A')
plt.scatter([1,2,3], [45,50,52], color = 'red', label='Class B')

plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Relationship between hours studied and exam score of two class')
plt.grid(True)
plt.legend(loc='upper left')
plt.show()
```

# Subplots

```python
#plt.subplot(nrows, ncols, index)
x = [1,2,3,4]
y = [10,20,15,25]

plt.subplot(1,2,1)
plt.plot(x,y,color='pink')
plt.title('Line chart')

plt.subplot(1,2,2)
plt.bar(x,y,color='red')
plt.title('Bar Chart')

plt.tight_layout()
plt.show()
```

```python
#fig, ax = plt.subplots(nrows, ncols, figsize=(width, height))
fig, ax = plt.subplots(1,2, figsize = (10,5))

x = [1,2,3,4]
y = [10,20,15,25]

ax[0].plot(x,y,color = 'black')
ax[0].set_title('Line Plot')

ax[1].bar(x,y,color='lightgreen')
ax[1].set_title('Bar Chart')

fig.suptitle('Comparison of line and bar chart')
plt.tight_layout()
plt.show()
```
