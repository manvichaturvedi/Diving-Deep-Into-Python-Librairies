# Pandas Data Cleaning
Data Cleaning Means Fixing Bad Data in your set.Bad data could be empty cels, null values, wrong data, duplicates etc

Empty cells can give you wrong result when you analyze data.So it is necessary to remove them.

```python
#removing null values
new_df = df.dropna()
new_df.info()
```

```python
#inplace = True is used to change the original dataframe
df.dropna(inplace=True)
```

```python
df.info()
```

# Replacing Empty Values
fillna() method is used to replace cells with a value

```python
old_df = pd.read_csv('titanic.csv')
old_df.head()
```

```python
old_df.info()
```

```python
#filling null values in age column with avg age
mean_age = old_df['Age'].mean()
old_df.fillna({"Age" : mean_age},inplace=True)
```

```python
old_df.info()
```

Insights : all null values in Age column is filled with mean age.

```python
x = old_df['Age'].mode()
x
```

```python
y = old_df['Age'].median()
y
```

```python
#removing duplicaes
old_df.drop_duplicates(inplace=True)
```

```python
old_df
```
